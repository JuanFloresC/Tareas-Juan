```python
# Importamos pandas y matplotlib. Son librerías que nos ayudan en el proceso
import pandas as pd
import matplotlib.pyplot as plt
#Definimos dataframes y llamamos al archivo csv
app = "/content/drive/MyDrive/Tarea 4/Tarea_04.csv"
df = pd.read_csv(app)

# Crear el gráfico de barras y definimos el ancho de las barras
fig, ax = plt.subplots()
bar_width = 0.35

# Definir la posición de las barras
r1 = df.index
r2 = [x + bar_width for x in r1]

# Graficar las barras. Definimos los colores delas barras, del fondo para diferenciarlas
plt.bar(r1, df['Uso total minutos'], color='skyblue', width=bar_width, edgecolor='grey', label='Uso total del teléfono (minutos)')
plt.bar(r2, df['Tiempo en la app en minutos'], color='salmon', width=bar_width, edgecolor='grey', label='Tiempo en la app más usada (minutos)')

# Agregar leyenda y título
plt.xlabel('Día', fontweight='bold', fontsize=15)
plt.xticks([r + bar_width for r in range(len(df))], df['Día'])
plt.legend()
plt.title('Comparación de uso del teléfono y tiempo en la app más usada por día')

# Mostrar el gráfico
plt.show()
```
